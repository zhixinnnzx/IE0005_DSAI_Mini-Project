```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
cardioData = pd.read_csv('cardio_train.csv')
cardioData.head()
```

```python
print("Data type : ", type(cardioData))
print("Data dims : ", cardioData.shape)
```

```python
height = pd.DataFrame(cardioData['height'])
print("Data type : ", type(height))
print("Data dims : ", height.size)
height.head()

#Viewing statistical distribution of Length
f = plt.figure(figsize=(24, 4))
sb.boxplot(data=height, orient = "h", color = "cornflowerblue")
f = plt.figure(figsize=(24, 12))
sb.histplot(data=height, color = "royalblue")
f = plt.figure(figsize=(24, 12))
sb.violinplot(data=height, orient='h')
```

```python
#Viewing ratios of smokers versus non-smokers in "Age" using catplot and describe method
smoke = pd.DataFrame(cardioData['smoke'])
sb.catplot(y = 'smoke', data = cardioData, kind = "count", height = 8)
smoke.describe()
```

```python
#Viewing ratios of cardio versus non-cardio in "Age" using catplot and describe method
cardio = pd.DataFrame(cardioData['cardio'])
sb.catplot(y = 'cardio', data = cardioData, kind = "count", height = 8)
smoke.describe()
```
